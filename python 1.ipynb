```python
# Data type in Python

# 1. int
# 2. float
# 3. str
# 4. bool
# 5. list
# 6. tuple
# 7. set
# 8. dict
# 9. None
# 10. complex
# 11. range
# 12. bytes
# 13. bytearray
```

```python
# Examples of each data type in Python

# 1. int
example_int = 10
print("int:", example_int)

# 2. float
example_float = 10.5
print("float:", example_float)

# 3. str
example_str = "Hello, World!"
print("str:", example_str)

# 4. bool
example_bool = True
print("bool:", example_bool)

# 5. list
example_list = [1, 2, 3, 4, 5]
print("list:", example_list)

# 6. tuple
example_tuple = (1, 2, 3, 4, 5)
print("tuple:", example_tuple)

# 7. set
example_set = {1, 2, 3, 4, 5}
print("set:", example_set)

# 8. dict
example_dict = {"name": "Alice", "age": 25}
print("dict:", example_dict)

# 9. None
example_none = None
print("None:", example_none)

# 10. complex
example_complex = 1 + 2j
print("complex:", example_complex)

# 11. range
example_range = range(5)
print("range:", list(example_range))

# 12. bytes
example_bytes = b"Hello"
print("bytes:", example_bytes)

# 13. bytearray
example_bytearray = bytearray(5)
print("bytearray:", example_bytearray)
```

```python
# float(): Converts a value into a floating-point number. For example:
num_str = "3.14"
num_float = float(num_str)
print(num_float)  # Output: 3.14
num_int = int(num_float)
print(num_int)
```

```python
# str(): Converts a value into a string. For example:
num_int = 42
num_str = str(num_int)
print(num_str)  # Output: "42"
print(type(num_str))
```

```python
data = "12345678910"
print(type(data))
# using list()
data_list = list(data)
print(data_list)

# using tuple()
data_tuple = tuple(data)
print(data_tuple)

# using set() 
data_set = set(data)
print(data_set)
```

```python
# bool(): Converts a value into a boolean. For example:
num = 42
bool_val = bool(num)
print(bool_val)  # Output: True
```

```python
# Example of type conversion for each datatype

# Integer to float
num_int = 10
num_float = float(num_int)
print(f"Integer to float: {num_float} (type: {type(num_float)})")

# Float to integer
num_float = 3.14
num_int = int(num_float)
print(f"Float to integer: {num_int} (type: {type(num_int)})")

# Integer to string
num_int = 42
num_str = str(num_int)
print(f"Integer to string: {num_str} (type: {type(num_str)})")

# String to integer
num_str = "123"
num_int = int(num_str)
print(f"String to integer: {num_int} (type: {type(num_int)})")

# String to float
num_str = "3.14"
num_float = float(num_str)
print(f"String to float: {num_float} (type: {type(num_float)})")

# List to tuple
data_list = [1, 2, 3]
data_tuple = tuple(data_list)
print(f"List to tuple: {data_tuple} (type: {type(data_tuple)})")

# Tuple to list
data_tuple = (1, 2, 3)
data_list = list(data_tuple)
print(f"Tuple to list: {data_list} (type: {type(data_list)})")

# List to set
data_list = [1, 2, 2, 3]
data_set = set(data_list)
print(f"List to set: {data_set} (type: {type(data_set)})")

# Set to list
data_set = {1, 2, 3}
data_list = list(data_set)
print(f"Set to list: {data_list} (type: {type(data_list)})")

# Integer to boolean
num_int = 0
bool_val = bool(num_int)
print(f"Integer to boolean: {bool_val} (type: {type(bool_val)})")

# String to boolean
str_val = ""
bool_val = bool(str_val)
print(f"String to boolean: {bool_val} (type: {type(bool_val)})")
```
